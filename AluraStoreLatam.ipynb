### Importación de datos



```python
import pandas as pd

url = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_1%20.csv"
url2 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_2.csv"
url3 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_3.csv"
url4 = "https://raw.githubusercontent.com/alura-es-cursos/challenge1-data-science-latam/refs/heads/main/base-de-datos-challenge1-latam/tienda_4.csv"

tienda = pd.read_csv(url)
tienda2 = pd.read_csv(url2)
tienda3 = pd.read_csv(url3)
tienda4 = pd.read_csv(url4)

tienda.head()
```

#1. Análisis de facturación



```python
ingreso_tienda_1 = tienda['Precio'].sum()
ingreso_tienda_2 = tienda2['Precio'].sum()
ingreso_tienda_3 = tienda3['Precio'].sum()
ingreso_tienda_4 = tienda4['Precio'].sum()
print("Ingreso total tienda 1:", ingreso_tienda_1)
print("Ingreso total tienda 2:", ingreso_tienda_2)
print("Ingreso total tienda 3:", ingreso_tienda_3)
print("Ingreso total tienda 4:", ingreso_tienda_4)

```

```python
ingreso_t1 = tienda["Precio"].sum()
ingreso_t2 = tienda2["Precio"].sum()
ingreso_t3 = tienda3["Precio"].sum()
ingreso_t4 = tienda4["Precio"].sum()

print("Ingresos totales:")
print(f"Tienda 1: ${ingreso_t1:,.0f}")
print(f"Tienda 2: ${ingreso_t2:,.0f}")
print(f"Tienda 3: ${ingreso_t3:,.0f}")
print(f"Tienda 4: ${ingreso_t4:,.0f}")
```

# 2. Ventas por categoría

```python
# Agrupar los datos por categoría y contar la cantidad de productos vendidos
ventas_categoria_tienda_1 = tienda.groupby('Categoría del Producto')['Producto'].count()
ventas_categoria_tienda_2 = tienda2.groupby('Categoría del Producto')['Producto'].count()
ventas_categoria_tienda_3 = tienda3.groupby('Categoría del Producto')['Producto'].count()
ventas_categoria_tienda_4 = tienda4.groupby('Categoría del Producto')['Producto'].count()

# Mostrar las categorías más populares por tienda
print("Ventas por categoría - Tienda 1:")
print(ventas_categoria_tienda_1)
print("\nVentas por categoría - Tienda 2:")
print(ventas_categoria_tienda_2)
print("\nVentas por categoría - Tienda 3:")
print(ventas_categoria_tienda_3)
print("\nVentas por categoría - Tienda 4:")
print(ventas_categoria_tienda_4)
```

```python
import matplotlib.pyplot as plt

# Agrupación de ventas por categoría en cada tienda
ventas_categoria_1 = tienda['Categoría del Producto'].value_counts()
ventas_categoria_2 = tienda2['Categoría del Producto'].value_counts()
ventas_categoria_3 = tienda3['Categoría del Producto'].value_counts()
ventas_categoria_4 = tienda4['Categoría del Producto'].value_counts()

# Visualización con gráficos de barras
plt.figure(figsize=(12, 8))

plt.subplot(2, 2, 1)
ventas_categoria_1.plot(kind='bar', color='skyblue')
plt.title("Ventas por Categoría - Tienda 1")
plt.ylabel("Cantidad de Ventas")
plt.xticks(rotation=45)

plt.subplot(2, 2, 2)
ventas_categoria_2.plot(kind='bar', color='lightgreen')
plt.title("Ventas por Categoría - Tienda 2")
plt.ylabel("Cantidad de Ventas")
plt.xticks(rotation=45)

plt.subplot(2, 2, 3)
ventas_categoria_3.plot(kind='bar', color='salmon')
plt.title("Ventas por Categoría - Tienda 3")
plt.ylabel("Cantidad de Ventas")
plt.xticks(rotation=45)

plt.subplot(2, 2, 4)
ventas_categoria_4.plot(kind='bar', color='orange')
plt.title("Ventas por Categoría - Tienda 4")
plt.ylabel("Cantidad de Ventas")
plt.xticks(rotation=45)

plt.tight_layout()
plt.show()
```

# 3. Calificación promedio de la tienda

```python
# Calificación promedio por tienda
calificaciones = [calificacion_promedio_tienda_1, calificacion_promedio_tienda_2, calificacion_promedio_tienda_3, calificacion_promedio_tienda_4]

# Gráfico de barras horizontales
plt.figure(figsize=(8, 5))
plt.barh(tiendas, calificaciones, color='darkorange')
plt.title("Calificación Promedio por Tienda")
plt.xlabel("Calificación Promedio")
plt.ylabel("Tiendas")
plt.xlim(0, 5)  # Limitar el rango de la calificación a 0-5
plt.show()
```

# 4. Productos más y menos vendidos

```python
import matplotlib.pyplot as plt

# Agrupar por Producto y contar las ventas de cada uno
ventas_producto_tienda_1 = tienda.groupby('Producto')['Producto'].count().sort_values(ascending=False)
ventas_producto_tienda_2 = tienda2.groupby('Producto')['Producto'].count().sort_values(ascending=False)
ventas_producto_tienda_3 = tienda3.groupby('Producto')['Producto'].count().sort_values(ascending=False)
ventas_producto_tienda_4 = tienda4.groupby('Producto')['Producto'].count().sort_values(ascending=False)

# Mostrar los productos más vendidos y menos vendidos por tienda
print("Productos más vendidos - Tienda 1:")
print(ventas_producto_tienda_1.head(1))
print("Productos menos vendidos - Tienda 1:")
print(ventas_producto_tienda_1.tail(1))

print("\nProductos más vendidos - Tienda 2:")
print(ventas_producto_tienda_2.head(1))
print("Productos menos vendidos - Tienda 2:")
print(ventas_producto_tienda_2.tail(1))

print("\nProductos más vendidos - Tienda 3:")
print(ventas_producto_tienda_3.head(1))
print("Productos menos vendidos - Tienda 3:")
print(ventas_producto_tienda_3.tail(1))

print("\nProductos más vendidos - Tienda 4:")
print(ventas_producto_tienda_4.head(1))
print("Productos menos vendidos - Tienda 4:")
print(ventas_producto_tienda_4.tail(1))

# Visualización de los productos más vendidos en cada tienda (top 10)
plt.figure(figsize=(10, 6))
plt.subplot(2, 2, 1)
ventas_producto_tienda_1.head(10).plot(kind='barh', title="Top 10 Productos más vendidos Tienda 1", color='skyblue')
plt.subplot(2, 2, 2)
ventas_producto_tienda_2.head(10).plot(kind='barh', title="Top 10 Productos más vendidos Tienda 2", color='lightgreen')
plt.subplot(2, 2, 3)
ventas_producto_tienda_3.head(10).plot(kind='barh', title="Top 10 Productos más vendidos Tienda 3", color='salmon')
plt.subplot(2, 2, 4)
ventas_producto_tienda_4.head(10).plot(kind='barh', title="Top 10 Productos más vendidos Tienda 4", color='lightcoral')

plt.tight_layout()
plt.show()
```

# 5. Envío promedio por tienda

```python
# Cálculo del costo de envío promedio
envio_promedio_1 = tienda['Costo de envío'].mean()
envio_promedio_2 = tienda2['Costo de envío'].mean()
envio_promedio_3 = tienda3['Costo de envío'].mean()
envio_promedio_4 = tienda4['Costo de envío'].mean()

# Lista de promedios y nombres de tiendas
envios_promedio = [envio_promedio_1, envio_promedio_2, envio_promedio_3, envio_promedio_4]
nombres_tiendas = ['Tienda 1', 'Tienda 2', 'Tienda 3', 'Tienda 4']

# Mostrar los valores en consola
print("Costo de envío promedio por tienda:")
for tienda, valor in zip(nombres_tiendas, envios_promedio):
    print(f"{tienda}: ${valor:,.2f}")

# Gráfico de barras con etiquetas
import matplotlib.pyplot as plt

plt.figure(figsize=(8, 5))
bars = plt.bar(nombres_tiendas, envios_promedio, color='mediumpurple')
plt.title("Costo de Envío Promedio por Tienda")
plt.xlabel("Tiendas")
plt.ylabel("Costo de Envío Promedio")
plt.grid(axis='y', linestyle='--', alpha=0.7)

# Agregar los valores arriba de cada barra
for bar in bars:
    yval = bar.get_height()
    plt.text(bar.get_x() + bar.get_width()/2.0, yval + 500, f"${yval:,.0f}", ha='center', va='bottom')

plt.tight_layout()
plt.show()


```
